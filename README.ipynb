## __Physiology rig__
Documentation of the physiology rig for head-fixed in vivo olfactory experiments on mice. Each section or subsection includes a list of the required parts and a description of the assembly. Links to product websites, STEP files or drawings are provided wherever necessary.


### __1. Temporal Odour Delivery Device (tODD)__
tODD (Ackels et al., 2021) can be used to reliably present olfactory stimuli ranging from simple odour pulses to complex high-frequency plume structures.

- [<u> to 6 mm connectors</u>](https://de.rs-online.com/web/p/pneumatische-anschlusse/9171007) for connecting the 6 mm tubes to the filters
- [<u>Pressure regulator</u>](https://de.rs-online.com/web/p/pneumatik-regler/7967323) and [<u>gauge</u>](https://de.rs-online.com/web/p/druckmessgerate/2353084)

- [<u>Flow meter</u>](https://ph.rs-online.com/web/p/flow-sensors/1982981) and [<u>thread adapter</u>](https://de.rs-online.com/web/p/pneumatische-anschlusse/0812235)
- Flow meter holders (STEP files: [<u>four-slot</u>](files/STEP_Files), five-slot) and [<u>10/32 screws</u>](https://de.rs-online.com/web/p/innensechskantschrauben/2454839)
- Thorlabs parts: [post holders](https://www.thorlabs.com/newgrouppage9.cfm?objectgroup_id=1268), [posts](https://www.thorlabs.de/newgrouppage9.cfm?objectgroup_id=1266) and [clamping forks](https://www.thorlabs.com/newgrouppage9.cfm?objectgroup_id=887) for mounting the four-slot flow meter holder on top of the five-slot holder (optional)
- [<u>Flow regulator</u>](https://de.rs-online.com/web/p/pneumatische-funktionsanschlusse/1975292?gb=s)

Air from a pressurised source (at 0.2 MPa) is passed through a water separator, mist separator and a pressure regulator (set to 0.05 MPa). The air coming out of the pressure regulator is split into eight input lines (flowing into the odour bottles) and a carrier line (for the carrier air stream), all of which are then connected to separate flow meters. The input lines are then connected to flow regulators while the carrier line is split into two lines for the two manifolds ([described later](#1c-valves-and-odour-delivery)).

The bottles are connected to custom PTFE lids mounted on custom brackets. O-rings are used to reduce the chances of leakage. The lids have two threaded holes - one for the flow regulator (input) and the second for the MINSTAC safety screen (output). A bracket holder holds the two brackets (i.e. eight odour bottles), one on top of the other. The holder is connected to a modified Thorlabs post with two M4 screws (the existing hole was expanded for the screw to go in, and an additional hole was drilled).
Teflon tubings are used to carry the output from the odour bottles. Detailed instructions for the tubing preparation and assembly can be found [<u>here</u>](https://www.theleeco.com/uploads/2021/04/EFS-9th-Edition-Oct-2019.pdf#page=188). There are, however, a couple of suggestions to make the procedure easier. While screwing the collet onto the tube end, the tubing needs to extend just beyond the point where the thread inside the collet ends and not all the way to the end of the collet. Also, the tubing grip is not necessary, and its use may leave marks on the tubes.

- [<u>Articulated arm</u>](https://www.conrad.de/de/p/mantona-mantona-befestigungsset-gopro-sony-actioncams-actioncams-1528133.html?hk=SEM&WT.mc_id=google_pla&gad_source=1&gclid=Cj0KCQjw7JOpBhCfARIsAL3bobd8YBPDQC31p1D3hLtEIUbH2Pu06UJ8J7vUc687ryKD0ksnQuqUHhAaAlYJEALw_wcB#attributesNotes_delivery) and [<u>adapter</u>]()

The outputs from the odour bottles are then directed to VHS (Very High Speed) valves mounted on the manifolds. The two manifolds, each with four valves, are inserted into a custom holder. The carrier lines are connected to one side of the manifolds through MINSTAC tubings. At the other end of the manifolds, a 3 boss manifold is used to combine the outputs from both of them. The 3 boss manifold is fixed to a custom holder, which is, in turn, connected to an articulated arm for flexible positioning of the odour delivery point. The valves are connected to a custom valve driver board and controlled with PulseBoy ([<u>described later</u>](#3-software-and-data-acquisition)) for odour presentation.

- Holders 1 and 2(STEP file and drawing 1 and 2)

The wheel is first fixed on the axle with an M4 screw to prevent lateral movement. The rotary encoder, attached to holder 2 with M3 screws, is connected to one end of the axle via a shaft coupler. The other end of the axle passes through holder 1 with a bearing. Axle length may be adjusted depending on the setup by cutting away the extra length. The rotary encoder is then connected to the rotary encoder module (following the manufacturers instructions). The module is powered via a USB-B cable, and analogue output can be read through a BNC cable connected to the screw terminals on the module.

The rig uses a Basler camera for pupillometry and a webcam [<u>without the IR filter</u>](https://youtu.be/eoMUJxUo4d8?si=8q_OPeFo1WALGWoY) as a body camera to monitor the animal during experiments. Both cameras are connected to the PC via USB. An additional GPIO cable with an M8 6-pin connector sends trigger inputs to the pupillometry camera. The open end of the GPIO cable is soldered to a BNC cable (check line specifications before soldering - use GND and I or I/O), which is then connected to a cDAQ (described later). An IR LED from Thorlabs is used for ambient light. The hot mirror and the pupillometry camera are aligned so that the camera acquires images via the mirror. Articulated arms (for cameras) and a ball and socket mount (for the mirror) are used for easier alignment.

A custom PCB is used to power and acquire analogue output from the sensors. A PID can be simultaneously used with the flow sensor to measure and calibrate odour concentrations and flow from the odour delivery device.

The overall workflow is designed to trigger each trial and data acquisition at the beginning of inhalation. Data acquisition units include two cDAQ-9174 chassis (cDAQ1 and cDAQ2)  and a CED Micro1401-4 with an expansion (together referred to as CED henceforth). Short descriptions of the software are given below.

- __Spike2__: data acquisition software for CED. It acquires data from the rotary encoder, PID and flow and respiration sensors. It is also used to send out a trigger pulse to PulseBoy at the beginning of each respiration cycle (by setting a threshold value for the analogue signal from the respiration sensor).

- __PulseBoy__: a Python-based GUI for controlling NI pulse trains. It operates VHS valves via the valve driver board to present odour stimuli. Trials are saved as trialbanks, and each trial can be triggered by trigger pulses from Spike2. Two extra channels are used to 1) trigger camera acquisition via Bonsai and 2) send a 10 KHz pulse train (clock) to CED/Spike2 as a reference frame for time.

- __Bonsai__: a visual reactive programming language. It triggers the pupillometry camera and acquires data from both cameras based on external triggers from PulseBoy. (Provide link to the workflow)

- __NI MAX__: configures NI hardware (cDAQs and corresponding modules).

cDAQ1 is connected with valves, AI and DIO modules. The valve module sends pulses to the valve driver board via [cable]. Channels 16 and 17 are used for clock and trigger, respectively, using SMA to BNC adapters. The trigger channel is further split with a BNC T-adapter to record the trigger pulses in Spike2. The AI module can be used as an analogue input device (it is not used for anything in the current setup, but the default configuration of PulseBoy requires an analogue device to be connected - could be changed if necessary). The DIO module receives trigger pulses from Spike2. Since Bonsai and PulseBoy cannot simultaneously access a cDAQ chassis, cDAQ2 with another DIO module is used for Bonsai to 1) send frame triggers (100 Hz) to the pupillometry camera and 2) receive trigger pulses from PulseBoy. The frame triggers are also recorded in Spike2 by splitting the output line with a BNC T-adapter.
